````
# Paso 1: Importar libreras
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
!pip install -q scipy
from scipy.stats import beta
````

````
# Paso 2: Cargar archivo con los riesgos
riesgos = 'riesgos.xlsx'
df=pd.read_excel(riesgos)
df.head()
````

````
# Paso 3: Definir cantidad de iteraciones
iteraciones =3000
````

````
# Paso 4: Funcin PERT robusta y con soporte para oportunidades

import numpy as np
from scipy.stats import beta

def distribucion_pert(minimo, mas_probable, maximo, lamb=4):
    """
    Retorna una distribucin beta-pert vlida y
    con soporte para oportunidades (valores negativos).
    """

    # Validar NaN  distribucin sin impacto
    if any(np.isnan([minimo, mas_probable, maximo])):
        return beta(1, 1, loc=0, scale=0)  # sin impacto

    # Ordenar correctamente los valores (funciona para amenazas y oportunidades)
    minimo, mas_probable, maximo = sorted([minimo, mas_probable, maximo])

    # Caso sin incertidumbre (mximo = mnimo)
    if minimo == maximo:
        return beta(1, 1, loc=minimo, scale=0)

    # Calcular media PERT
    mean = (minimo + lamb * mas_probable + maximo) / (lamb + 2)

    denominador = (mas_probable - mean) * (maximo - minimo)
    if denominador == 0:
        return beta(a=2, b=2, loc=minimo, scale=maximo - minimo)

    # Calcular parmetros alpha y beta
    alpha = ((mean - minimo) * (2 * mas_probable - minimo - maximo)) / denominador
    beta_param = alpha * (maximo - mean) / (mean - minimo)

    # Validacin de parmetros
    if any([np.isnan(alpha), np.isnan(beta_param), alpha <= 0, beta_param <= 0]):
        return beta(a=2, b=2, loc=minimo, scale=maximo - minimo)

    # Retornar distribucin Beta-PERT
    return beta(a=alpha, b=beta_param, loc=minimo, scale=maximo - minimo)
````

````
# Paso 5: Simulacin Monte Carlo
resultados_simulacion = []
for _ in range(iteraciones):
    impacto_total = 0.0
    for _, riesgo in df.iterrows():
        probabilidad = float(riesgo["Probabilidad (%)"])

        # Normalizar: si la probabilidad viene como 50 en vez de 0.5, dividir por 100
        if probabilidad > 1:
            probabilidad = probabilidad / 100.0

        # Saltar NaN
        if np.isnan(probabilidad):
            continue

        if np.random.rand() <= probabilidad:
            dist = distribucion_pert(
                riesgo["Impacto (Min.)"],
                riesgo["Impacto (M.P.)"],
                riesgo["Impacto (Max.)"]
            )
            impacto = float(dist.rvs(size=1)[0])
            impacto_total += impacto

    resultados_simulacion.append(impacto_total)

resultados_simulacion = np.array(resultados_simulacion)

# Paso 6: Estadsticas para visualizacin
media = np.mean(resultados_simulacion)
p80  = np.percentile(resultados_simulacion, 80)
p90  = np.percentile(resultados_simulacion, 90)
````

````
# Paso 7: Se indican los puntos y se calculan las probabilidades
umbral_1 = 0
umbral_2 = 71100

# Calcular probabilidades acumuladas
prob_1 = np.mean(resultados_simulacion <= umbral_1)
prob_2 = np.mean(resultados_simulacion <= umbral_2)
````

````
# Paso 8: Texto con los resultados obtenidos
print("Reserva de Contingencia del Proyecto (Simulacin Monte Carlo):\n")
print(f"* Media del impacto: ${media:,.0f}")
print(f"* P80 (percentil 80): ${p80:,.0f}")
print(f"* P90 (percentil 90): ${p90:,.0f}\n")
print(f"* Probabilidad de que NO se requiera reserva de contingencia (impacto  ${umbral_1:,.0f}): {prob_1*100:.2f}%")
print(f"* Probabilidad de que una reserva de ${umbral_2:,.0f} sea suficiente (impacto  ${umbral_2:,.0f}): {prob_2*100:.2f}%\n")

# Paso 9: Grficos con los resultados obtenidos
# Preparar grfico con doble eje: histograma + CDF (Cumulative Distribution Function)
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

sns.set(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(12, 6))

# Grfico 1: Histograma
color_hist = "green"
sns.histplot(resultados_simulacion, bins=50, color=color_hist, ax=ax1)
ax1.set_xlabel("Impacto ($)")
ax1.set_ylabel("Frecuencia", color=color_hist)
ax1.tick_params(axis='y', labelcolor=color_hist)
ax1.grid(False)
ax1.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))  # Eje X con separador de miles

# Lneas de referencia (media, percentiles, umbrales)
ax1.axvline(media, color='red', linestyle='--', label=f'Media: ${media:,.0f}')
ax1.axvline(p80, color='orange', linestyle='--', label=f'P80: ${p80:,.0f}')
ax1.axvline(p90, color='green', linestyle='--', label=f'P90: ${p90:,.0f}')
ax1.axvline(umbral_1, color='purple', linestyle=':', label=f'Umbral 1: ${umbral_1:,.0f} ({prob_1*100:.1f}%)')
ax1.axvline(umbral_2, color='brown', linestyle=':', label=f'Umbral 2: ${umbral_2:,.0f} ({prob_2*100:.1f}%)')

# Grfico 2: Porcentaje acumulado - CDF
ax2 = ax1.twinx()
color_cdf = "blue"
sorted_values = np.sort(resultados_simulacion)
cdf_values = np.arange(1, len(sorted_values) + 1) / len(sorted_values)
ax2.plot(sorted_values, cdf_values, color=color_cdf, linewidth=2, label="CDF (Acumulado)")
ax2.set_ylabel("Probabilidad acumulada", color=color_cdf)
ax2.tick_params(axis='y', labelcolor=color_cdf)
ax2.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1))  # Eje Y secundario en porcentaje
ax2.set_yticks(np.arange(0, 1.05, 0.05))  # Intervalos de 5%

# Ttulo y leyenda
plt.title("Reserva de Contingencia del Proyecto (Simulacin de Monte Carlo)\n"
          "Histograma, CDF (Funcin de Distribucin Acumulada) y Umbrales")
fig.legend(loc="center right", bbox_to_anchor=(0.9, 0.5))

plt.tight_layout()

# OPCIONAL: Exportar la imagen (puedes cambiar el formato: png, jpeg, pdf)
plt.savefig("histograma.png")

plt.show()
````

````
# Paso 10: Generar tabla de Frecuencias

import math
import pandas as pd

# Identificar el mnimo y mximo del rango de resultados
min_val = resultados_simulacion.min()
max_val = resultados_simulacion.max()
rango = max_val - min_val

# Definir la unidad de clase adecuada con base en la magnitud del rango
# Se busca que haya entre ~20 y 50 clases
unidad_clase = 10 ** int(math.floor(math.log10(rango / 30)))

# Redondear lmites
limite_inferior = unidad_clase * math.floor(min_val / unidad_clase)
limite_superior = unidad_clase * math.ceil(max_val / unidad_clase)

# Crear clases (se agrega una clase extra porque np.arange no incluye el ltimo valor por defecto)
bins = np.arange(limite_inferior, limite_superior + unidad_clase, unidad_clase)
clases = pd.IntervalIndex.from_breaks(bins, closed='right')
# Intervalos cerrados a la derecha: (1000, 2000] significa >1000 hasta 2000 inclusive

# Calcular la frecuencia absoluta y el porcentaje acumulado
frecuencia = pd.cut(resultados_simulacion, bins=clases).value_counts().sort_index()

porcentaje_acumulado = (frecuencia.cumsum() * 100 / frecuencia.sum()).round(2).astype(str) + '%'

# Definir los valores de clase como el lmite superior (right) del intervalo
valor_clases = [int(interval.right) for interval in clases]

# Crear tabla con resultados
tabla_histograma = pd.DataFrame({
    "Clase": valor_clases[:len(frecuencia)],
    "Frecuencia": frecuencia.values,
    "% Acumulado": porcentaje_acumulado.values
})

# Formatear solo la columna "Clase" con separador de miles
tabla_histograma["Clase"] = tabla_histograma["Clase"].apply(lambda x: f"{x:,}")

# Mostrar tabla
print("Tabla Histograma:\n")
print(tabla_histograma.head(20))  # Muestra primeras 20 filas

# OPCIONAL: Exportar a Excel
tabla_histograma.to_excel("tabla_histograma.xlsx", index=False)
````

````
# ===== Paso 11: Sensibilidad (Tornado OAT) usando ID como etiqueta =====

def simular_totales(riesgos_df, iteraciones=3000, forzar=None):
    """
    Simula el impacto total para 'iteraciones' corridas.
    - riesgos_df: columnas ['Probabilidad (%)','Impacto (Min.)','Impacto (M.P.)','Impacto (Max.)']
    - forzar: tuple (idx, modo) con modo  {'min','mp','max'} para fijar ese riesgo.
    """
    tot = np.empty(iteraciones, dtype=float)

    for i in range(iteraciones):
        total = 0.0
        for j, (_, r) in enumerate(riesgos_df.iterrows()):
            prob = float(r["Probabilidad (%)"])
            if prob > 1:  # por si viniera 50 en vez de 0.50
                prob = prob / 100.0
            if np.isnan(prob):
                continue

            if np.random.rand() <= prob:
                if forzar and j == forzar[0]:
                    if forzar[1] == "min":
                        a = b = c = float(r["Impacto (Min.)"])
                    elif forzar[1] == "mp":
                        a = b = c = float(r["Impacto (M.P.)"])
                    else:  # 'max'
                        a = b = c = float(r["Impacto (Max.)"])
                else:
                    a = float(r["Impacto (Min.)"])
                    b = float(r["Impacto (M.P.)"])
                    c = float(r["Impacto (Max.)"])
                dist = distribucion_pert(a, b, c)
                total += float(dist.rvs(1)[0])

        tot[i] = total
    return tot

# Opcional: filtrar filas completamente vlidas para el tornado
mask_valid = df[["Probabilidad (%)","Impacto (Min.)","Impacto (M.P.)","Impacto (Max.)"]].notna().all(axis=1)
df_tornado = df.loc[mask_valid].reset_index(drop=True)

# Simulacin base (P80)
base_tot = simular_totales(df_tornado, iteraciones=iteraciones)
base_p80 = np.percentile(base_tot, 80)

# Sensibilidad OAT (LOW/HIGH) por riesgo
labels = df_tornado["ID"].astype(str).tolist()   # <--- SOLO ID en el eje Y
delta_low, delta_high = [], []

for idx in range(len(df_tornado)):
    tot_low  = simular_totales(df_tornado, iteraciones=iteraciones, forzar=(idx, "min"))
    tot_high = simular_totales(df_tornado, iteraciones=iteraciones, forzar=(idx, "max"))
    p80_low  = np.percentile(tot_low,  80)
    p80_high = np.percentile(tot_high, 80)
    delta_low.append(p80_low - base_p80)       # izquierda
    delta_high.append(p80_high - base_p80)     # derecha

# Orden de mayor a menor impacto absoluto (tornado)
orden = np.argsort(np.maximum(np.abs(delta_low), np.abs(delta_high)))
labels     = [labels[i]     for i in orden]
delta_low  = [delta_low[i]  for i in orden]
delta_high = [delta_high[i] for i in orden]

# Grfico Tornado
fig, ax = plt.subplots(figsize=(10, 5))
y = np.arange(len(labels))
ax.barh(y, delta_low,  color="navy",      label="LOW (Min)")
ax.barh(y, delta_high, color="goldenrod", label="HIGH (Max)")
ax.axvline(0, color="black", linewidth=1)

ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.set_xlabel(" P80 del impacto total (vs. caso base)")
ax.set_title("Sensibilidad por riesgo (Tornado OAT sobre P80)  Etiquetas: ID")
ax.legend()
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))

plt.tight_layout()
plt.show()

print(f"P80 base: {base_p80:,.0f}")
````
